```python
import pandas as pd

file_path = '/content/drive/MyDrive/Colab Notebooks/Upgrad_LR/Colleges.xlsx'
try:
    df = pd.read_excel(file_path)
    print("DataFrame loaded successfully:")
    display(df.head())
except FileNotFoundError:
    print(f"Error: The file was not found at {file_path}")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
df = df.drop('Unnamed: 0', axis=1)
display(df.head())
```

```python
df.describe().round(2)
```

```python
df = pd.get_dummies(df, columns=['Private'], drop_first=True, dtype=int)
display(df.head())
```

```python
y = df['Grad.Rate']
X = df.drop('Grad.Rate', axis=1)
display(X.head())
display(y.head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("Training set shape (X_train, y_train):", X_train.shape, y_train.shape)
print("Testing set shape (X_test, y_test):", X_test.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

# Initialize the Linear Regression model
model = LinearRegression()

# Fit the model to the training data
model.fit(X_train, y_train)

print("Linear Regression model trained successfully.")
```

```python
import pandas as pd

# Output model coefficients
coefficients = pd.DataFrame(model.coef_, X.columns, columns=['Coefficient'])
display(coefficients.round(2))
```

```python
# Filter out the categorical predictor 'Private_Yes'
non_categorical_coefficients = coefficients.drop('Private_Yes')

# Find the predictor with the highest absolute coefficient among the non-categorical predictors
highest_coefficient_predictor = non_categorical_coefficients.abs().idxmax()

print(f"The non-categorical predictor with the highest coefficient is: {highest_coefficient_predictor.values[0]}")
```

```python
# Obtain predictions
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

print("Predictions obtained for training and testing sets.")
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np

# Evaluate on training set
r2_train = r2_score(y_train, y_train_pred)
mse_train = mean_squared_error(y_train, y_train_pred)
mae_train = mean_absolute_error(y_train, y_train_pred)

print("Training Set Evaluation:")
print(f"R-squared: {r2_train:.2f}")
print(f"Mean Squared Error: {mse_train:.2f}")
print(f"Mean Absolute Error: {mae_train:.2f}")

print("\n" + "="*30 + "\n")

# Evaluate on testing set
r2_test = r2_score(y_test, y_test_pred)
mse_test = mean_squared_error(y_test, y_test_pred)
mae_test = mean_absolute_error(y_test, y_test_pred)

print("Testing Set Evaluation:")
print(f"R-squared: {r2_test:.2f}")
print(f"Mean Squared Error: {mse_test:.2f}")
print(f"Mean Absolute Error: {mae_test:.2f}")
```

```python
# Get the intercept term and round it to 2 decimal places
intercept = round(model.intercept_, 2)

print(f"The intercept term is: {intercept}")
```
