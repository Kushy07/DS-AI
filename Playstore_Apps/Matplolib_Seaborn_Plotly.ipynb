```python
plt.figure(figsize = [10,5])
df3.groupby(['Updated_month'])['Rating'].mean().plot()
plt.show()
```

```python
monthly = pd.pivot_table(df3, index = "Updated_month", values="Installs", columns = "Content Rating", aggfunc=sum)
```

```python
monthly.plot(kind = "bar", stacked = "True", figsize = [10,6])
plt.show()
```

```python
# Get proportions
monthly_perc = monthly[["Everyone","Everyone 10+","Mature 17+","Teen"]].apply(lambda x: x/x.sum(), axis = 1)
```

```python
monthly_perc.plot(kind = "bar", stacked = "True", figsize = [10,6])
plt.show()
```

```python
res = df3.groupby(['Updated_month'])[['Rating']].mean()
res.reset_index(inplace=True)
```
